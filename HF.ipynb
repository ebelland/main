## Conductivity
In literature it is possible to find different values for $σ_{HF}$ [4],[5],[6].

Literature data confirm an almost linear relation for diluted solutions.
The conductivity for diluted $H_{2}SiF_{6}$ has not been found in literature; therefore, it has been extrapolated from concentrated solutions data of [6].

For low $H_{2}SiF_{6}$ concentrations the trend is linear and the linear coefficient has also a linear behavior with temperature therefore $σ_{H_{2}SiF_{6}}$ at 21°C is approximated to 48.49mS/cm.

```python
#@title <i>Concentraction and conducivity from volumes</i> { run: "auto", vertical-output: true, display-mode: "form" }

def GetConcentration(DIWv:float, HFv:float, NH4Fv:float, HFin:float, NH4Fin:float, Temperature:float):
  TOTv=HFv+NH4Fv+DIWv
  HFn=1000.0*HFv/TOTv
  NH4Fn=1000.0*NH4Fv/TOTv
  DIWn=1000.0*DIWv/TOTv
  HFw=HFn*rbf(Temperature,HFin)#(1.0+(1.16-1.0)*HFin/49.0)
  NH4Fw=NH4Fn*(1.0+(1.11-1.0)*NH4Fin/40.0)
  DIWw=DIWn*dWaters(Temperature)
  TOTw=HFw+NH4Fw+DIWw
  HFc=HFw/TOTw*HFin
  NH4Fc=NH4Fw/TOTw*NH4Fin
  DIWc=DIWw/TOTw*100.0
  return HFc,NH4Fc,DIWc

def getConductivity(cNH4F:float,cHF:float):
  sN=sNH4F(cNH4F)
  if isnan(sN):
      sN=0.0
  sH=sHF(cHF)
  return sH+sN

def ErHFO(Concentration:float, Temperature):
  Er0=Ers(Concentration,extrapolate=True)
  Ea=EaOi(Concentration)  
  return Er0*np.exp(1000.0*(Temperature-23.0)*Ea/(8.3144598*(Temperature+273.16)*296.16))

def ErHF(Concentration:float, Temperature):
  Ea=Eas(Concentration,extrapolate=True)
  Er0=Ers(Concentration,extrapolate=True)
  return Er0*np.exp(1000.0*(Temperature-23.0)*Ea/(8.3144598*(Temperature+273.16)*296.16))
  
#@markdown HF incoming concentration (%)
HFin =  49#@param {type:"number"}
#@markdown HF volume or flow (l or l/m)
HFv =  202#@param {type:"number"}
#@markdown NH<sub>4</sub> incoming concentration (%) (for BOE) [optional]
NH4Fin =  40#@param {type:"number"}
#@markdown NH<sub>4</sub>F volume or flow (l or l/m) (for BOE) [optional]
NH4Fv =  0#@param {type:"number"}
#@markdown DIW volume or flow (l or l/m)
DIWv =  1798#@param {type:"number"}
#@markdown HF temperature (°C)
Temperature = 21.5 #@param {type:"number"}
#@markdown Etch target (Å) [optional]
EtchTarget = 300 #@param {type:"number"} 
t="<code style=color:black;font-size:18px;border:10px>"
HFc,NH4Fc,DIWc=GetConcentration(DIWv,HFv,NH4Fv,HFin,NH4Fin,Temperature)
t+=f'HF concentration:<b>{HFc:04.2f}</b>%<p>'
t+=f'NH<sub>4</sub>F concentration:<b>{NH4Fc:04.2f}</b>%<p>'
t+=f'H<sub>2</sub>O concentration:<b>{DIWc:04.2f}</b>%<p>'
s=getConductivity(NH4Fc,HFc)
t+=f'Conductivity:<b>{s:04.2f}</b>Ms/cm<p>'
t+=f'HF probe reading:<b>{sHFinv(s):04.2f}</b>%<p>'
Concentration=HFc
if NH4Fv>0:
  t+="Etch rate data not available for BOE"  
elif  Temperature>5 and Temperature<80 and Concentration>0.00001 and Concentration<40:
  er=ErHF(Concentration,Temperature)
  t+=f'Etch rate:<b>{er:04.2f}</b> Å/min<p>'
  if (EtchTarget>0) and (er>0):
    td=dt.timedelta(seconds=60.0*EtchTarget/er)
    t+=f'Dip time:<b>{td.seconds // 3600:02}:{td.seconds % 3600 // 60:02}:{td.seconds % 60:02}</b> hh:mm:ss<p>'
  erO=ErHFO(Concentration,Temperature)
  #t+=f'Etch rate Ohmin:<b>{erO:04.2f}</b> Å/min<p>'
else:
  t+="N:A."
t+="</code>"
display(HTML(t))
```

# $SiO_2$ etch rate calculation 
##Theory
The etch rate at a given temperature can be obtained using the Arrhenius equation $k=k_0 e^{-\frac{E_a}{RT}}$:

$$E_r(c,T)=E_r(c,T_{ref})\exp\frac{1000(T−T_{ref})E_a(c)}{R(T+273.16)(T_{ref}+273.16)}$$

Where $R=8.3144598$ is the gas constant.



```python
#@title <i>Etch rate calculation</i> { run: "auto", vertical-output: true, display-mode: "form" }
wHF=db.at[2,'Weight']
wSiO2=db.at[15,'Weight']
dSiO2=db.at[15,'Density']
wH2O=db.at[9,'Weight']
wH2SiF6=db.at[16,'Density']

#@markdown HF concentration (%)
Concentration =  1#@param {type:"number"}
#@markdown HF temperature (°C)
Temperature = 22.5 #@param {type:"number"}
#@markdown Etch target (Å) [optional]
EtchTarget = 90 #@param {type:"number"} 
t="<code style=color:black;font-size:18px;border:10px>"
if Temperature>5 and Temperature<80 and Concentration>0.00001 and Concentration<40:
  er=ErHF(Concentration,Temperature)
  t+=f'Etch rate:<b>{er:04.2f}</b> Å/min<p>'
  if (EtchTarget>0) and (er>0):
    td=dt.timedelta(seconds=60.0*EtchTarget/er)
    t+=f'Dip time:<b>{td.seconds // 3600:02}:{td.seconds % 3600 // 60:02}:{td.seconds % 60:02}</b> hh:mm:ss<p>'
  erO=ErHFO(Concentration,Temperature)
  #t+=f'Etch rate Ohmin:<b>{erO:04.2f}</b> Å/min<p>'

else:
  t+="N:A."
t+="</code>"
display(HTML(t))
```

```python
#@title <i>Etch rate adjustment</i> { run: "auto", vertical-output: true, display-mode: "form" }

def ErAdjustHF(Concentration:float, CurrentEtchRate:float, DesiredEtchRate:float):
  Ea=Eas(Concentration,extrapolate=True)
  Er0=Ers(Concentration,extrapolate=True)
  C1=Ea*1000.0/(8.3144598*296.16)
  C2=np.log(CurrentEtchRate/Er0)
  C3=np.log(DesiredEtchRate/Er0)
  TGuess=(23.0*C1+273.16*C2)/(C1-C2)
  TSet=(23.0*C1+273.16*C3)/(C1-C3)
  return TGuess,TSet

#@markdown HF concentration (%)
Concentration =  1#@param {type:"number"}
#@markdown Current etch rate (Å/min)
CurrentEtchRate =  45#@param {type:"number"}
#@markdown Desired etch rate (Å/min)
DesiredEtchRate =  46#@param {type:"number"}
t="<code style=color:black;font-size:18px;border:10px>"
TGuess,TSet=ErAdjustHF(Concentration,CurrentEtchRate,DesiredEtchRate)
t+=f'Estimated bath temperature:<b>{TGuess:04.2f}</b>°C<p>'
t+=f'Set bath temperature:<b>{TSet:04.2f}</b>°C<p>'
t+=f'{"Increase" if TSet>TGuess else "Decrease"} bath temperature by:<b>{np.abs(TSet-TGuess):04.2f}</b>°C<p>'
t+="</code>"
display(HTML(t))
```

# Recirculation system
In a recirculation reactor, the etch rate shift due to HF consumption after processing a certain number of wafers depends on the mass consumed by the reaction (in grams) and on the correction due to HF evaporation (in grams/hour): 

<br>$$\Delta{HF}=6m_{SiO_2} \cdot M_{HF}-\Delta{HF}^{Evap} \cdot t$$

<br>$$m_{SiO_2}=s\cdot n_W\cdot h\frac{\frac{D}{2}^2 d_{SiO_2}}{10^6 M_{SiO_2}}$$<p>

Where:<br>
$m_{SiO_2}$ is the mass of $SiO_2$ consumed during the reaction.<p>
$s$ are the number of oxidized wafer sides: 1 if the oxide is present only on the frontside (or backside) and 2 if the oxide is on both wafer sides.
$n_W$ is the number of processed wafers.<p>
$h$ is the oxide thickness in Å.<p>
$D$  is the wafer diameter in mm.<p>
$d_{SiO_2}$ is the $SiO_2$ density in g/cm\^3.<p> 
$M_{HF}$ and $M_{H_2SiF_6}$ are the molar masses of HF and exhafluorosilicic acid.<p>
$t$  is the bath lifetime.<p>
$\Delta{HF}^{Evap}$ is the amount of HF evaporated (g/hour) and depends on several factors, including bath temperature, tank surface and flow over the tank itself.<p>
The final HF concentration is given by:

$$[{HF}]_{fin}=[{HF}]_{init} - \frac{\Delta_{HF}}{10V}.$$<p>

Where:<p> 
$V$  is the tank volume (l).<p>
The amount of byproducts (exhaflurosilicic acid) generated during the reaction is:
$$\Delta{H_2SiF_6}=m_{SiO_2} \cdot M_{H_2SiF_6} \,(g)$$<p>

```python
#@title *Simulation* { run: "auto", display-mode: "form" }

def calcHFRecirc(Temperature,Concentration,Time,Lag,Wafers,Sides,Size,Runs,Evaporation,Volume,StdBy,Sel):
  out=pd.DataFrame(index=np.arange(0, Runs+1),columns=('Run', 'Concentration', 'Etch Rate','Temperature','Etch','Byproducts'))
  SiO2area=float(Wafers*Sides*Size*Size/400.0)*np.pi #cm² 'area dell'ossido
  dH2O=dWaters(Temperature)
  mHF=Concentration/100.0*(Volume*1000.0*dH2O)/wHF
  mH2SiF6=0.0
  cf=Concentration
  temp=Temperature
  tf=0.0
  er=0.0   
  for id in range(Runs+1):
    cf=mHF*wHF/(Volume*1000.0*dH2O)*100.0        
    er=ErHF(cf,temp)*Sel
    tf=er*float(Time+Lag)/60.0 #Å
    out.loc[id]= {'Run': id,'Concentration':cf, 'Etch Rate':er,'Temperature':temp,'Etch':tf,'Byproducts':mH2SiF6*wH2SiF6}
    mSiO2r=tf*1E-8*SiO2area*dSiO2/wSiO2  #moli di SiO2 attaccata
    mH2SiF6+=mSiO2r
    mHF*=(1.0+Evaporation/100.0/3600.0*float(Time+Lag+StdBy*60))
    mHF-=6.0*mSiO2r #moli di HF consumate
  return out

#@markdown *Bath temperature (°C)*
Temperature = 65 #@param {type:"number"}
#@markdown HF concentration (%)
Concentration = 5 #@param {type:"number"}
#@markdown *Dip time (s)*
Time = 62 #@param {type:"integer"}
#@markdown *Dip lag time (s)*
Lag = 0 #@param {type:"integer"}
#@markdown *Evaporation rate(%/hour)*
Evaporation = 0.01 #@param {type:"number"}
#@markdown This is a correction factor. A rate of 100% means that after one hour the concentration is doubled.

#@markdown *Oxidized sides*
TwoSides = True #@param {type:"boolean"}
#@markdown *Wafer size (mm)*
Size = 200 #@param[150,200,300]
#@markdown *Number of wafers per run*
Wafers = 25 #@param {type:"integer"}
#@markdown *Number of runs*
Runs =   1#@param {type:"integer"}
#@markdown *Tank volume (l)*
Volume =  20 #@param {type:"number"}
#@markdown *Time between runs (minutes)*
StdBy =  0#@param {type:"integer"} 
#@markdown *Selectivity (1:x)*
Selectiviy =  1#@param {type:"number"} 

Sides = 2 if TwoSides==True else 1

dfHF=calcHFRecirc(Temperature,Concentration,Time,Lag,Wafers,Sides,Size,Runs,Evaporation,Volume,StdBy,Selectiviy)

o1=[
    ["Etch Rate Initial",f"{dfHF['Etch Rate'].iloc[0]:4.2f}","Å/min"],   
    ["Etch Rate Final",f"{dfHF['Etch Rate'].iloc[Runs]:4.2f}",'Å/min'],   
    ["Concentration Initial",f"{dfHF['Concentration'].iloc[0]:4.2f}",'%'],   
    ["Concentration Final",f"{dfHF['Concentration'].iloc[Runs]:4.2f}",'%'],  
  ]

p1 = figure(tools=TOOLS,y_axis_type="linear",x_axis_type="linear",plot_height=300,plot_width=300)
p1.line(dfHF['Run'],dfHF['Etch Rate'],line_color='blue',line_width=2)
p1.yaxis.axis_label = 'Etch Rate (Å/min)'
p1.xaxis.axis_label = 'Runs'

p2 = figure(tools=TOOLS,y_axis_type="linear",x_axis_type="linear",plot_height=300,plot_width=300)
p2.line(dfHF['Run'],dfHF['Byproducts'],line_color='blue',line_width=2)
p2.yaxis.axis_label = 'Byproducts (g)'
p2.xaxis.axis_label = 'Runs'

p3 = figure(tools=TOOLS,y_axis_type="linear",x_axis_type="linear",plot_height=300,plot_width=300)
p3.line(dfHF['Run'],dfHF['Concentration'],line_color='blue',line_width=2)
p3.yaxis.axis_label = 'Concentration (%)'
p3.xaxis.axis_label = 'Runs'

p4 = figure(tools=TOOLS,y_axis_type="linear",x_axis_type="linear",plot_height=300,plot_width=300)
p4.line(dfHF['Run'],dfHF['Etch'],line_color='blue',line_width=2)
p4.yaxis.axis_label = 'Etched Oxide (Å)'
p4.xaxis.axis_label = 'Runs'

display_table(o1)
show(gridplot([[p1,p3],[p2,p4]]))
del dfHF
```

# Injection system

## Simple model
In case HF is injected into a process tank in a continuos mode, such as in DNS ONB tanks, the mixing between the water and the chemical can be considered ideal, therefore the concentration can be considered homogeneous inside the bath. 
In this case concentration can be estimated as:
$$C(t)=C_{in}(1-e^{-\frac{t}{\tau}})$$
$$\tau=\frac{V}{F}$$
where:<p>
$C(t)$ is the bath concentration as a function of time<p>
$C_{in}$ is the incoming concentration<p>
$\tau_{i}$ is the time constant, given by the ratio of the flow and tank volume<p> $V$ is the tank volume (litres)<p>
$F$ is the incoming flow (litres/min)<p>

During the dip time the concentration is supposed to be constant while during the rinse the concentration decreases according to:<p>
$$C(t)=C_d\,e^{-\frac{t}{\tau_r}}$$<p>
where $C_d$ is the concentration at the end of the dip time and $\tau_r$ is the time constant of the rinse step, that may have a different flow rate.<p>
The amount of etched oxide can be computed by numerically integrating the etch rate:<p>
$$E(t)=\int_{0}^{t_{tot}} E_r(C(t),T(t)) \,dt$$<p>
Where $t_{tot}$ is the time at the end of the rinse cycle.<p>


```python
#@title Simulation { run: "auto", vertical-output: false, display-mode: "form" }

def calcHFInjSimple(Temperature,HF_Concentration,HF_Flow,Time,InjectionTime,RinseTime,Water_Flow,Rinse_Flow,Volume):
  #time interval, smaller is higher resolution
  dt=0.1 #seconds
  #compute the number of iterations
  ninj,ndip,nrinse=int(InjectionTime/dt),int(Time/dt),int(RinseTime/dt)
  numberOfRows=ninj+nrinse+ndip #resize the ouput panda
  out=pd.DataFrame(index=np.arange(0, numberOfRows),columns=('Time', 'Concentration', 'Etch Rate','Temperature','Etch'))
  #injection phase
  temp=Temperature
  Total_Flow=Water_Flow+HF_Flow/1000.0
  CHFIn=(HF_Concentration/100.0*HF_Flow/1000.0)/Total_Flow*100.0  #injected HF Concentration

  cf=0.0
  mhf=0.0
  mh2o=0.0
  er=0.0
  tf=0.0
  tau=Volume/Total_Flow*60.0
  t=0.0
  for id in range(ninj):
    er=ErHF(cf,temp)
    tf+=er*dt/60.0
    cf=CHFIn*(1.0-np.exp(-t/tau))  
    out.loc[id]= {'Time': t,'Concentration':cf, 'Etch Rate':er,'Temperature':temp,'Etch':tf}
    t+=dt 

  for id in range(ninj,ninj+ndip):  
    er=ErHF(cf,temp)
    tf+=er*dt/60.0
    out.loc[id]= {'Time': t,'Concentration':cf, 'Etch Rate':er,'Temperature':temp,'Etch':tf}
    t+=dt 
  
  c0=cf
  t0=t
  tau=Volume/Rinse_Flow*60.0
  for id in range(ninj+ndip,numberOfRows):  
    cf=c0*np.exp(-(t-t0)/tau)
    er=ErHF(cf,temp)
    tf+=er*dt/60.0
    out.loc[id]= {'Time': t,'Concentration':cf, 'Etch Rate':er,'Temperature':temp,'Etch':tf}
    t+=dt
   
  return out

#@markdown *Bath temperature (C)*
Temperature = 22.0 #@param {type:"number"}
#@markdown HF facility concentration (%)
HF_Concentration = 49 #@param {type:"number"}
#@markdown HF injection flow (cc/min)
HF_Flow = 30000 #@param {type:"number"}
#@markdown Water Injection Flow rate (l/min)
Water_Flow = .01 #@param {type:"number"}
#@markdown Water Rinse Flow rate (l/min)
Rinse_Flow = 30 #@param {type:"number"}
#@markdown *Dip time (s)*
Time =  1#@param {type:"number"}
#@markdown *Injection time (s)*
InjectionTime = 300 #@param {type:"number"}
#@markdown *Rinse time (s)*
RinseTime = 1200 #@param {type:"number"}
#@markdown *Tank volume (l)*
Volume =  56#@param {type:"number"}

phf=calcHFInjSimple(Temperature,HF_Concentration,HF_Flow,Time,InjectionTime,RinseTime,Water_Flow,Rinse_Flow,Volume)
Total_Flow=Water_Flow+HF_Flow/1000.0
CHFIn=(HF_Concentration/100.0*HF_Flow/1000.0)/Total_Flow*100.0  #injected HF Concentration
maxConc=max(phf['Concentration'])
lastConc=phf.iloc[-1]['Concentration']

o1=[
    ["Max etch rate",format(max(phf['Etch Rate']),'4.2f'),'Å/min'], 
    ["Max concentration",format(maxConc,'4.2f'),'%'],
    ["Residual concentration",format(lastConc*100*1E6,'4.2f'),'ppm'],
    ["Relative max concentration",format(100.0*maxConc/CHFIn,'4.2f'),'%'],
    ["Etch",format(max(phf['Etch']),'4.2f'),'Å'],  
    ["Process time",datetime.timedelta(seconds=max(phf['Time'])),''] 
   ]
vline = Span(location=InjectionTime, dimension='height', line_color='red', line_width=1)
rline = Span(location=InjectionTime+Time, dimension='height', line_color='red', line_width=1)

p1 = figure(tools=TOOLS,y_axis_type='linear',x_axis_type='linear',plot_height=400,plot_width=600)
p1.line(phf['Time'],phf['Concentration'],line_color='blue',line_width=2)
p1.yaxis.axis_label = 'Concentration (%)'
p1.xaxis.axis_label = 'Time (s)'
p1.renderers.extend([vline,rline])

p2 = figure(tools=TOOLS,y_axis_type='linear',x_axis_type='linear',plot_height=400,plot_width=600)
p2.line(phf['Time'],phf['Etch Rate'],line_color='blue',line_width=2)
p2.yaxis.axis_label = 'Etch Rate (Å/m)'
p2.xaxis.axis_label = 'Time (s)'
p2.renderers.extend([vline,rline])

p3 = figure(tools=TOOLS,y_axis_type='linear',x_axis_type='linear',plot_height=400,plot_width=600)
p3.line(phf['Time'],phf['Etch'],line_color='blue',line_width=2)
p3.yaxis.axis_label = 'Etch (Å)'
p3.xaxis.axis_label = 'Time (s)'
p3.renderers.extend([vline,rline])

display_table(o1)
show(gridplot([[p1,p2],[p3]]))
del phf
```
